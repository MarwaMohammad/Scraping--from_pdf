```python
import pandas as pd
import re
from tabula import read_pdf
```

```python
dfs = read_pdf("data.pdf", multiple_tables=True)
```

```python
df = dfs[0]
df= df[6:]
df.reset_index(drop=True)
```

```python
df=df.iloc[:,[0,1,2,3,4]]
```

```python
ldc = []
Unterschrank = []
Kurzbezeichnung = []
untershrank=[]
onetow_ls = []
Punkte_ls=[]
three_eight_ls = []
nine_ten_ls = []
comment_ls = []
for i in range(0, df.shape[0]):
    untershrank_dc_old ={}
    if i in [1,11,21,27,33,43,50,56]:
        untershrank.append(str(df.iloc[i,0])+" " +str(df.iloc[i+1,0])+" " +str(df.iloc[i+2,0]))
    else:
        untershrank.append(None)
        
    Kurzbezeichnung.append(df.iloc[i,1]) 
    
    if i in [8,18,40]:
        comment_ls.append(str(df.iloc[i,2])+" " +str(df.iloc[i+1,2]))
        onetow_ls.append("nan")
    elif i in [9,19,41]:
        comment_ls.append(None)
        onetow_ls.append("nan")

        
    else:
        comment_ls.append(None)
        onetow_ls.append(df.iloc[i,2])
        
    three_eight_ls.append(df.iloc[i,3])
    nine_ten_ls.append(df.iloc[i,4])
```

```python
new_df = pd.DataFrame()
new_df['untershrank'] = untershrank
new_df = new_df.fillna(method='ffill')
new_df.untershrank = new_df.untershrank.str.replace("nan", "")
new_df.iloc[0,0] = "1 durchgehende Tür, 1 Einlegeboden"
```

```python
new_df['Kurzbezeichnung'] = Kurzbezeichnung
```

```python
Kurzbezeichnung_ls,Breite_ls,Bestell_Nr_ls =[], [], []
for i in range(new_df.shape[0]):
    if str(new_df.Kurzbezeichnung[i]) != "nan":
        Kurzbezeichnung, Breite, Bestell_Nr = new_df.Kurzbezeichnung[i].split(" ")
        Kurzbezeichnung_ls.append(Kurzbezeichnung)
        Breite_ls.append(Breite)
        Bestell_Nr_ls.append(Bestell_Nr)
        
    else:
        Kurzbezeichnung_ls.append(None)
        Breite_ls.append(None)
        Bestell_Nr_ls.append(None)

```
